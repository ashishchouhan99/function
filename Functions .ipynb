```python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(x):
  if x == 1 or x == 2 :
    return " Its prime "
  if x < 1 or x > 200 :
    return " Invalid data "
  for i in range(2 , x):
    if x % i == 0 :
      return f"{x} is not Prime"
    return f"{x} is Prime"

print(is_prime(100))
```

```python
# Q5. Create an iterator function in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci_iterator(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b


for number in fibonacci_iterator(10):
    print(number)
```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent

def generate(exp):
  for i in range (1,exp):
    yield 2**i

x = generate(10)
next(x)
next(x)
next(x)
next(x)
```

```python
# Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

temp_list=[10,0,15,273]
a=list(map(lambda x : (9/5)*(x)+32,temp_list))
print([round(i,2) for i in a])
```
